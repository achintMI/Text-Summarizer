```python
sentences = open('test.txt', 'r').read().decode('utf8')
```

```python
text = sentences.lower()
print(text)
```

```python
NEG_CONTRACTIONS = [
    (r'aren\'t', 'are not'),
    (r'can\'t', 'can not'),
    (r'couldn\'t', 'could not'),
    (r'daren\'t', 'dare not'),
    (r'didn\'t', 'did not'),
    (r'doesn\'t', 'does not'),
    (r'don\'t', 'do not'),
    (r'isn\'t', 'is not'),
    (r'hasn\'t', 'has not'),
    (r'haven\'t', 'have not'),
    (r'hadn\'t', 'had not'),
    (r'mayn\'t', 'may not'),
    (r'mightn\'t', 'might not'),
    (r'mustn\'t', 'must not'),
    (r'needn\'t', 'need not'),
    (r'oughtn\'t', 'ought not'),
    (r'shan\'t', 'shall not'),
    (r'shouldn\'t', 'should not'),
    (r'wasn\'t', 'was not'),
    (r'weren\'t', 'were not'),
    (r'won\'t', 'will not'),
    (r'wouldn\'t', 'would not'),
    (r'ain\'t', 'am not') # not only but stopword anyway
]
```

```python
import re
for t in NEG_CONTRACTIONS:
    sentences = re.sub(t[0], t[1], text)
```

```python
tokens = tokenize(sentences)
```

```python
OTHER_CONTRACTIONS = {
    "'m": 'am',
    "'ll": 'will',
    "'s": 'has', # or 'is' but both are stopwords
    "'d": 'had'  # or 'would' but both are stopwords
}
```

```python
tokens = [OTHER_CONTRACTIONS[token] if OTHER_CONTRACTIONS.get(token) else token for token in tokens]
```

```python
from nltk.corpus import movie_reviews, stopwords
BLACKLIST_STOPWORDS = ['over','only','very','not','no']
ENGLISH_STOPWORDS = set(stopwords.words('english'))
remove_punc = r'[a-z]+'
tokens = [word for word in tokens if re.search(remove_punc, word)]
tokens = [token for token in tokens if token not in ENGLISH_STOPWORDS]
```

```python
from nltk.stem import PorterStemmer
stemmer = nltk.PorterStemmer()
```

```python
for i in tokens:
    print(i)
```

```python
tokens = [stemmer.stem(word) for word in tokens]
```
